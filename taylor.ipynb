```python
#importing modules
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sympy import *
```

```python
# func - the function you would like to approximate.
# start, the beginning of your interval.
# end, the end of your interval.
# degree, the last integer where you will be truncating your series.
# fixed_c, the point you will be expanding around.

def TaylorSeriesApprox (func,start,end,degree,fixed_c):
    x1 = np.linspace(start,end,100) # independent x vaules
    tsa=0 
    for i in range(degree): #taylor series expanding for the given specification
        tsa += (diff(func,x,i).subs(x,fixed_c))/(factorial(i))*(x-fixed_c)**i
    
    #substituting the x values to find y 
    y1 = [tsa.subs(x,j) for j in x1[:]]
    return y1
```

```python
#x taken as variable
x=sym.symbols("x")

#the function you would like to approximate.
func = x*sin(x)**2+cos(x)

# independent x vaules
x1 = np.linspace(-10,10,100)

#Approximated y vlaues
y1 = TaylorSeriesApprox (func,-10,10,99,0)

#subsituted y values
y2 = [func.subs(x,j) for j in x1[:]]

#plotting the grpah
plt.plot(x1,y1,'ro')
plt.plot(x1,y2,'k-')
plt.xlabel('$x$',fontsize=12)
plt.ylabel('$xsin^2(x)+cos(x)$',fontsize=12)
plt.legend(['Taylor Approximation','Actual'],loc=2)
plt.show()
```
